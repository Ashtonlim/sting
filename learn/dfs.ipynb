```python
def intToRoman(num):
    seq = [ 1,4,5,9,10,40,50,90,100,400,500,900,1000, float('inf') ]
    mapping = {
        1:'I',
        4:'IV',
        5:'V',
        9:'IX',
        10:'X',
        40:'XL',
        50:'L',
        90:'XC',
        100:'C',
        400:'CD',
        500:'D',
        900:'CM',
        1000:'M',
        float('inf'):'INF'
    }


    numeral = ''
    cur, next = len(seq)-1, len(seq)-2
    while next >= 0:
        if seq[cur] > num > seq[next]:
            numeral += seq[next]
        else:
            cur -= 1
            next -= 1
    return numeral

print(intToRoman(3))
```
